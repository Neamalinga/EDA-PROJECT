** Outlier Analysis**

```python
outlier_summary = []

for col in selected_features:

    q1 = bank_data[col].quantile(0.25)
    q3 = bank_data[col].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = ((bank_data[col] < lower) | (bank_data[col] > upper)).sum()

    outlier_summary.append({
        "feature": col,
        "outlier_count": outliers,
        "lower_bound": lower,
        "upper_bound": upper
    })

outlier_df = pd.DataFrame(outlier_summary)

outlier_df["percentage"] = (outlier_df["outlier_count"] / len(bank_data)) * 100

outlier_df.sort_values("percentage", ascending=False)
```

 For the majority of non-bankrupt companies, their ROA and Net Income fall within a typical, healthy range. However, the outliers often represent a subset of companies that are exceptionally profitable, highly efficient in using their assets, or have unusually high net income. These are the 'super-performers' that exist on the positive extreme of the distribution.

Healthy companies ideally generate sufficient cash internally or raise equity. Companies heading for bankruptcy, however, often cannot secure equity or generate enough internal funds, forcing them into a continuous cycle of borrowing to meet obligations or fund operations. This creates an extreme 'borrowing dependency,' where they are excessively reliant on lenders, which is an unsustainable position and a strong precursor to financial collapse.
In essence, these outliers for bankrupt companies are not just statistical anomalies; they are critical red flags in financial statements. They indicate that these companies have reached extreme levels of financial imbalance, where debt and short-term obligations have become unmanageable relative to their assets and ability to generate profits. These extreme values are precisely what a bankruptcy prediction model needs to identify to effectively classify companies at high risk.



These outliers represent extreme financial states crucial for bankruptcy prediction: unusually high ROA/Net Income for non-bankrupt firms indicate 'super-performers', while exceptionally high debt and borrowing dependency highlight critical distress in bankrupt companies. Retaining them provides the model with vital information on the full spectrum of financial health and distress.

## INSIGHTS

**Higher Debt Ratios** Indicate Higher Risk: Debt ratio % shows a strong positive correlation (0.25). This means companies with a higher proportion of debt relative to their total assets are significantly more likely to be bankrupt. It's a key indicator of financial leverage and potential distress.

**Increased Current Liabilities**  **Current Liability to Assets** (0.19) and Current Liability to Current Assets (0.17) both have positive correlations. This suggests that companies struggling with a higher proportion of short-term obligations relative to their assets or current assets are at a greater risk of bankruptcy, as they might face liquidity issues.

**Borrowing Dependency** is a Red Flag: Borrowing dependency (0.18) correlates positively with bankruptcy. A higher reliance on external borrowing to fund operations often indicates that a company is unable to generate sufficient funds internally, a classic sign of impending financial collapse.

 **Net Income to Total Assets** shows the strongest negative correlation (-0.32). Companies that generate higher net income relative to their total assets are far less likely to go bankrupt.

**Return on Assets (ROA)** is Critical: All variants of Return on Assets (ROA(A)..., ROA(B)..., ROA(C)...) exhibit significant negative correlations (around -0.26 to -0.28). This consistently demonstrates that companies with lower returns on their assets (meaning they are less efficient at using their assets to generate profits) are at a much higher risk of bankruptcy. These metrics are core indicators of financial health
